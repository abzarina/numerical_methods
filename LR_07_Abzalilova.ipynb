```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_experiments = 60     
std_deviation = 3   
true_k = 0.5    
true_b = 2    

x_values = np.array(range(num_experiments))
true_function = np.array([true_k * z + true_b for z in range(num_experiments)])
observed_values = true_function + np.random.normal(0, std_deviation, num_experiments)

mean_x = x_values.sum() / num_experiments
mean_y = observed_values.sum() / num_experiments
a2 = np.dot(x_values.T, x_values) / num_experiments
a11 = np.dot(x_values.T, observed_values) / num_experiments

calculated_k = (a11 - mean_x * mean_y) / (a2 - mean_x ** 2)
calculated_b = mean_y - calculated_k * mean_x
calculated_function = np.array([calculated_k * z + calculated_b for z in range(num_experiments)])

plt.scatter(x_values, observed_values, s=2, c='red', label='Наблюдения')
plt.plot(true_function, label='Истинная зависимость')
plt.plot(calculated_function, c='red', label='МНК')
plt.grid(True)
plt.legend()
plt.show()
```
